## <i> i.) Import Libraries and Dependencies </i>

```python
# Import libraries and dependencies
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pathlib import Path
import os
import re
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

## <i> ii.) Prepare the DataFrames </i>

```python
# Get filename from dir and store as 'algoData_asset' for each individual asset file to use later in labelling/filenaming

# Create empty list for all .csv result files in algoData_ANN_results
filelist = []
assetNames = []

# Create dataframe list & dictionary to hold unique dataframe variable names 
df_list = []
d = {}

dirlist = []
dirlist = os.listdir('../Datasets/algoData_ANN_results/')

# Create filelist of all '_results.csv' ANN result files
for file in dirlist:
    if '.csv' in file:
        filelist.append(file)
        
        filename = Path(f'../Datasets/algoData_ANN_results/{file}')
        os.path.basename(filename)
        os.path.dirname(filename)
        os.path.splitext(filename)
        os.path.splitext(os.path.basename(filename))
        algoData_asset = os.path.splitext(os.path.basename(filename))
        algoData_asset = algoData_asset[0] 
        algoData_asset = algoData_asset.split('_')
        algoData_asset = algoData_asset[0]
        assetNames.append(algoData_asset)
        
        d['df_%s'%file] = pd.read_csv(filename, 
                index_col='Date', 
                parse_dates=False,
                infer_datetime_format=True)
        d['df_%s'%file].index = pd.to_datetime(d['df_%s'%file].index, format="%Y-%m-%d", utc=True)
        d['df_%s'%file].index = d['df_%s'%file].index.normalize()
        d['df_%s'%file].columns = [col_name+'_{}'.format(algoData_asset) for col_name in d['df_%s'%file].columns]
        df_list.append(d['df_%s'%file])
        
# Print filelist and all assetNames available
print(f'Filelist: {filelist}')
print()
print(f'AssetNames: {assetNames}')
print()


# Uniquely re-name dataframes and uniquely index them in dictionary
a = {}
for i in range(len(assetNames)):
    a['df_%s'%assetNames[i]] = df_list[i]

print(f'Unique dataframes: {a.keys()}')
```

```python
# Concatenate all dataframes
# where -> (a = dictionary of all individual asset dataframes)
temp_df= []
for i in a.keys():
    df = a[i]
    
    temp_df.append(df)
all_assets_df = pd.concat(temp_df, axis=1)
```

```python
# Cumulative Returns for All Assets 

from matplotlib.pyplot import figure

# Filter 'all_assets_df' to only include the 'cumANNStrategyReturns' columns values 
gen_cols_1 = all_assets_df.columns[all_assets_df.columns.str.contains("cumANNStrategyReturns_")]
gen_cols = gen_cols_1.tolist()

# Set title label of plot & title properties
label = "Artificial Neural Network - Cumulative Strategy Returns [ReLU/Sigmoid Activation] (Test Period)"
csfont = {'fontname':'Arial', 'fontsize': 18, 'fontweight': 'normal'}
axis_font = {'fontname':'Arial', 'fontsize': 16, 'fontweight': 'normal'}

# Plot all_assets_df 
# Matplotlib Style Types (style.use) - 'seaborn-white', 'fivethirtyeight', 'ggplot', 'bmh'
all_assets_df.plot(y=gen_cols, figsize=(16,9))
plt.style.use('seaborn-white')
plt.legend(labels=assetNames, prop={'size': 14})
plt.title(label, fontdict=None, loc="center", pad=10, y=None, **csfont)
plt.grid(color='grey', linestyle='-', which='both', linewidth=0.4, axis='both')
plt.xlabel("Date", **axis_font)
plt.ylabel("Cumulative Returns", **axis_font)
plt.savefig('../Datasets/algoData_ANN_results/ann_strat_returns_test.png')
plt.figure(figsize=(50,20))
```

```python
# Create descriptive statistics function for assets

def descriptiveStats(df, initial, riskFree, assetName, period):

    # Calculate Descriptive Statistics
    
    start_date = df.index.min()
    end_date = df.index.max()

    start = str(start_date.day)+'-'+str(start_date.month)+'-'+str(start_date.year)
    end = str(end_date.day)+'-'+str(end_date.month)+'-'+str(end_date.year)

    days = (end_date - start_date).days
    years = days/365

    init_investment = initial
    rf = riskFree

    buyHold_start = init_investment
    ann_start = init_investment
    
    buyHold_end = round(df.iloc[:,3][-1] * init_investment,2)
    ann_end = round(df.iloc[:,3][-1] * init_investment,2)
    
    buyHold_max_dailyReturn = round(df.iloc[:,1].max(),6)
    ann_max_dailyReturn = round(df.iloc[:,-1].max(),6)
    
    buyHold_min_dailyReturn = round(df.iloc[:,1].min(),6)
    ann_min_dailyReturn = round(df.iloc[:,-1].min(),6)
    
    buyHold_max_drawdown = round(((df.iloc[:,3].min() - df.iloc[:,3].max())/df.iloc[:,3].max()),6)
    ann_max_drawdown = round(((df.iloc[:,3].min() - df.iloc[:,3].max())/df.iloc[:,3].max()),6)
        
    buyHoldSignals = ((df.iloc[:,3] > 0) | (df.iloc[:,3] < 0)).count()
    annSignals = (df.iloc[:,0] == 1) | (df.iloc[:,0] == -1).count()
    annSignals = annSignals.count()
    
    buyHoldPos = (df.iloc[:,1] > 0).value_counts()[True]
    annPos = (df.iloc[:,2] > 0).value_counts()[True]
    
    
    buyHoldPosPerc = round((buyHoldPos/buyHoldSignals),6)
    annPosPerc = round((annPos/annSignals),6)
    
    buyHoldPosSum = df.where(df.iloc[:,1] > 0).sum()
    buyHoldPosSum = buyHoldPosSum[2]
    annPosSum = df.where(df.iloc[:,2] > 0).sum()
    annPosSum = annPosSum[2]
    
    buyHoldPosAvg = round((buyHoldPosSum/buyHoldPos),6)
    annPosAvg = round((annPosSum/annPos),6)
    
    buyHoldNeg = (df.iloc[:,1] < 0).value_counts()[True]
    annNeg = (df.iloc[:,2] < 0).value_counts()[True]
    
    buyHoldNegPerc = round((buyHoldNeg/buyHoldSignals),6)
    annNegPerc = round((annNeg/annSignals),6)
    
    buyHoldNegSum = df.where(df.iloc[:,1] < 0).sum()
    buyHoldNegSum = buyHoldNegSum[2]
    annNegSum = df.where(df.iloc[:,2] < 0).sum()
    annNegSum = annNegSum[2]
    
    buyHoldNegAvg = round((buyHoldNegSum/buyHoldNeg),6)
    annNegAvg = round((annNegSum/annNeg),6)
    
    buyHold_annualReturn = df.iloc[:,1].apply(lambda x: (1+x)).cumprod().iloc[-1]**(1/years) - 1
    ann_annualReturn = df.iloc[:,2].apply(lambda x: (1+x)).cumprod().iloc[-1]**(1/years) - 1
    
    buyHold_annualVol = df.iloc[:,1].apply(lambda x: np.log(1+x)).std()*np.sqrt(252)
    ann_annualVol = df.iloc[:,2].apply(lambda x: np.log(1+x)).std()*np.sqrt(252)
    
    buyHold_Sharpe = round((buyHold_annualReturn-rf)/buyHold_annualVol,2)
    ann_Sharpe = round((ann_annualReturn-rf)/ann_annualVol,2)
    
    buyHold_variance = df.iloc[:,1].var()
    
    ann_covariance = df.iloc[:,2].cov(df.iloc[:,1])
    
    buyHold_beta = 1
    
    ann_beta = round(ann_covariance/buyHold_variance,3)
    

    # Table of Descriptive Statistics

    head = [f'<b>Statistic for {assetName} <b>', '<b>Buy&Hold<b>', '<b>ANN<b>']
    labels = ['<b>Start Date<b>', '<b>End Date<b>','<b>Initial Investment<b>', '<b>Ending Investment<b>','--------------------------',
              '<b>Signals<b>', '<b>Winning Trades<b>', '<b>Losing Trades<b>', '<b>% Winning<b>', '<b>% Losing<b>', 
              '<b>Average Profit<b>','<b>Average Loss<b>', '--------------------------', 
              '<b>Max Daily Return<b>','<b>Min Daily Return<b>', '<b>Max Drawdown<b>', '<b>Annual Return<b>',
              '<b>Annual Volatility<b>', '<b>Sharpe Ratio<b>', '<b>Beta<b>']


    buyHold_stats = [start, end,'${:,}'.format(buyHold_start), '${:,}'.format(buyHold_end), '--------------------------',
                     buyHoldSignals, buyHoldPos, buyHoldNeg,'{:.2%}'.format(buyHoldPosPerc),'{:.2%}'.format(buyHoldNegPerc), 
                     '{:.2%}'.format(buyHoldPosAvg), '{:.2%}'.format(buyHoldNegAvg), '--------------------------', 
                     '{:.2%}'.format(buyHold_max_dailyReturn), '{:.2%}'.format(buyHold_min_dailyReturn), 
                     '{:.2%}'.format(buyHold_max_drawdown), '{:.2%}'.format(buyHold_annualReturn), 
                     '{:.2%}'.format(buyHold_annualVol), buyHold_Sharpe, buyHold_beta]


    ann_stats = [start, end,'${:,}'.format(ann_start), '${:,}'.format(ann_end),'--------------------------', 
                  annSignals, annPos, annNeg,'{:.2%}'.format(annPosPerc),'{:.2%}'.format(annNegPerc),
                  '{:.2%}'.format(annPosAvg), '{:.2%}'.format(annNegAvg), '--------------------------', 
                  '{:.2%}'.format(ann_max_dailyReturn), '{:.2%}'.format(ann_min_dailyReturn), 
                  '{:.2%}'.format(ann_max_drawdown), '{:.2%}'.format(ann_annualReturn), 
                  '{:.2%}'.format(ann_annualVol), ann_Sharpe, ann_beta]

    fig11 = go.Figure(data=[go.Table(
        header=dict(values=head,
                fill_color='paleturquoise',
                align='left'),
        cells=dict(values=[labels, buyHold_stats, ann_stats],
            fill_color='lavender',
            align='left'))
    ])

    fig11.update_layout(margin=dict(l=0, r=0, b=0,t=0), width=950, height=650)
    
    
    fileName = assetName+'_descriptiveStatistics_'+period+'.png'
    
    fig11.write_image(f'../Datasets/algoData_ANN_results/{fileName}')

    return fig11.show()
```
